```python
from mabel.data import Reader, STORAGE_CLASS
from mabel.adapters.disk import DiskReader
```

```python
tweets = Reader(
    inner_reader=DiskReader,
    dataset="data/tweets",
    raw_path=True,
    persistence=STORAGE_CLASS.MEMORY)
```

```python
tweets.count()
```

```python
tweets.distinct().count()
```
